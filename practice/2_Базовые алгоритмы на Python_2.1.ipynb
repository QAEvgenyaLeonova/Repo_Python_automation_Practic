```python
import math

def square(n):
    return math.ceil(n * n)

kva = int(input('count: '))
result = square(kva)
print(result)
```

```python
def fizz_buzz(n):
    if n % 3 == 0 and n % 5 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    else:
        print(n)

fizz_buzz(15)
```

```python
def month_to_season(num_month):
    if 3 <= num_month <= 5:
        return 'Весна'
    elif 6 <= num_month <= 7:
        return 'Лето'
    elif 8 <= num_month <= 9:
        return 'Осень'
    else:
        return 'Зима'

mes_num_user = int(input('Номер '))
print(month_to_season(mes_num_user))

```
