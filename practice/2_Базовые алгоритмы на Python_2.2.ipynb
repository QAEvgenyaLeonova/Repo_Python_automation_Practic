```python
s = 0
for show in range(11):
    s += show
    print(s, end=' ')
```

```python
s = 'asdfghjklzxcvbnm'
for show in range(0, len(s), 3):
    print(s[show], end=' ')
```

```python
s = 'Колбаса'
for show in range(len(s)):
    s = s[1:] + s[:1]
    print(s, end=' ')
```
